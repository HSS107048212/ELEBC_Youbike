```python
import pandas as pd

# Function to calculate IN_ride and OUT_ride for a single row of values
def calculate_rides(values):
    differences = values[:-1] - values[1:]
    IN_ride = differences[differences > 0].sum()
    OUT_ride = differences[differences < 0].sum()
    return IN_ride, OUT_ride

# General function to process multiple rows and update the DataFrame
def process_all_values(data):
    # Adding new columns for IN_ride and OUT_ride
    data['IN_ride'] = 0
    data['OUT_ride'] = 0

    for index, row in data.iterrows():
        values = row[1:-2].values  # Update to exclude new columns from calculation
        IN_ride, OUT_ride = calculate_rides(values)
        data.at[index, 'IN_ride'] = IN_ride
        data.at[index, 'OUT_ride'] = OUT_ride

    return data

# Call the function to process all rows and update the DataFrame
updated_data = process_all_values(data)
updated_data
```
